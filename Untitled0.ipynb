### Dataset Construction

We parse image metadata and XML annotations to create a structured dataframe containing image paths, class labels, and bounding-box coordinates for model training and evaluation.

```python
def parse_xml(xml_path):
    tree = ET.parse(xml_path)
    root = tree.getroot()

    filename = root.find("filename").text

    width = int(root.find("size/width").text)
    height = int(root.find("size/height").text)

    boxes = []

    for obj in root.findall("object"):
        bbox = obj.find("bndbox")

        xmin = int(bbox.find("xmin").text)
        ymin = int(bbox.find("ymin").text)
        xmax = int(bbox.find("xmax").text)
        ymax = int(bbox.find("ymax").text)

        boxes.append([xmin, ymin, xmax, ymax])

    return filename, width, height, boxes
```

```python
sample_file = os.path.join(xml_dir, xml_files[0])

filename, width, height, boxes = parse_xml(sample_file)

print("Filename:", filename)
print("Image Width:", width)
print("Image Height:", height)

print("\nBounding Boxes:")
for box in boxes:
    print(box)
```

### Classification Data Generators

Separate generators are created for lightweight models (128×128) and transfer learning models (224×224).

```python
def make_flow(gen, df_, size):
    return gen.flow_from_dataframe(
        dataframe=df_,
        x_col="image_path",
        y_col="label",
        target_size=size,
        batch_size=BATCH_SIZE,
        class_mode="categorical",
        classes=CLASSES,
        shuffle=True
    )

def make_val(df_, size):
    return plain_gen.flow_from_dataframe(
        dataframe=df_,
        x_col="image_path",
        y_col="label",
        target_size=size,
        batch_size=BATCH_SIZE,
        class_mode="categorical",
        classes=CLASSES,
        shuffle=False
    )
```

```python
tr128 = make_flow(
    aug_gen,
    train_df,
    IMG_SIZE
)

v128 = make_val(
    val_df,
    IMG_SIZE
)

te128 = make_val(
    test_df,
    IMG_SIZE
)

print("128x128 generators ready")
```

```python
tr224 = make_flow(
    aug_gen,
    train_df,
    IMG_SIZE_TL
)

v224 = make_val(
    val_df,
    IMG_SIZE_TL
)

te224 = make_val(
    test_df,
    IMG_SIZE_TL
)

print("224x224 generators ready")
```

# 6. Localization Preprocessing

The localization model requires bounding-box coordinates as regression targets. Bounding boxes are normalized and custom generators are created for simultaneous classification and localization.

```python
def normalise_boxes(df_in):

    df = df_in.copy()

    x1n, y1n, x2n, y2n = [], [], [], []

    for _, row in df.iterrows():

        img = cv2.imread(row["image_path"])

        if img is None:

            x1n.append(0)
            y1n.append(0)
            x2n.append(0)
            y2n.append(0)

            continue

        h, w = img.shape[:2]

        x1n.append(row["xmin"] / w if pd.notna(row["xmin"]) else 0)
        y1n.append(row["ymin"] / h if pd.notna(row["ymin"]) else 0)
        x2n.append(row["xmax"] / w if pd.notna(row["xmax"]) else 0)
        y2n.append(row["ymax"] / h if pd.notna(row["ymax"]) else 0)

    df["x1_n"] = x1n
    df["y1_n"] = y1n
    df["x2_n"] = x2n
    df["y2_n"] = y2n

    return df
```

# 7. Shared Training Utilities

To ensure consistent training and evaluation across all models, common callbacks and utility functions are defined and reused throughout the project.

```python
def make_callbacks(
    name,
    monitor="val_accuracy",
    patience=5
):

    return [

        EarlyStopping(
            monitor=monitor,
            patience=patience,
            restore_best_weights=True,
            verbose=1
        ),

        ReduceLROnPlateau(
            monitor=monitor,
            factor=0.5,
            patience=3,
            min_lr=1e-6,
            verbose=1
        ),

        ModelCheckpoint(
            f"{name}_best.h5",
            monitor=monitor,
            save_best_only=True,
            verbose=0
        )

    ]

print("Callbacks ready.")
```

```python
all_results = {}
all_histories = {}

print("Result containers initialized.")
```

```python
def eval_cls_model(
    model,
    test_flow,
    model_name
):

    y_true = []
    y_pred = []

    for i in range(len(test_flow)):

        X, Y = test_flow[i]

        preds = model.predict(
            X,
            verbose=0
        )

        y_true.extend(
            np.argmax(Y, axis=1)
        )

        y_pred.extend(
            np.argmax(preds, axis=1)
        )

    y_true = np.array(y_true)
    y_pred = np.array(y_pred)

    acc = accuracy_score(
        y_true,
        y_pred
    )

    f1 = f1_score(
        y_true,
        y_pred,
        average="macro"
    )

    print(f"\n--- {model_name} ---")
    print(f"Accuracy : {acc:.4f}")
    print(f"Macro F1 : {f1:.4f}")

    print(
        classification_report(
            y_true,
            y_pred,
            target_names=CLASSES
        )
    )

    all_results[model_name] = (
        y_true,
        y_pred
    )

    return y_true, y_pred
```

```python
def plot_history(history, model_name):

    plt.figure(figsize=(12,4))

    # Accuracy
    plt.subplot(1,2,1)

    plt.plot(
        history.history["accuracy"],
        label="Train"
    )

    plt.plot(
        history.history["val_accuracy"],
        label="Validation"
    )

    plt.title(f"{model_name} Accuracy")

    plt.xlabel("Epoch")

    plt.ylabel("Accuracy")

    plt.legend()

    # Loss
    plt.subplot(1,2,2)

    plt.plot(
        history.history["loss"],
        label="Train"
    )

    plt.plot(
        history.history["val_loss"],
        label="Validation"
    )

    plt.title(f"{model_name} Loss")

    plt.xlabel("Epoch")

    plt.ylabel("Loss")

    plt.legend()

    plt.tight_layout()

    plt.show()
```

# 8. Model 1 – Artificial Neural Network (ANN)

The Artificial Neural Network (ANN) serves as the baseline model for this study. Images are flattened into one-dimensional feature vectors and passed through fully connected layers for multi-class classification.

**Key Features**
- Baseline deep learning model
- Fully connected architecture
- ReLU activation
- Dropout regularization
- Softmax output layer

```python
ann_model = tf.keras.Sequential([

    layers.Flatten(
        input_shape=(IMG_SIZE[0], IMG_SIZE[1], 3)
    ),

    layers.Dense(
        512,
        activation="relu"
    ),

    layers.Dropout(0.3),

    layers.Dense(
        256,
        activation="relu"
    ),

    layers.Dropout(0.3),

    layers.Dense(
        NUM_CLASSES,
        activation="softmax"
    )
])

ann_model.compile(
    optimizer="adam",
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

ann_model.summary()
```

```python
ann_history = ann_model.fit(

    tr128,

    validation_data=v128,

    epochs=EPOCHS,

    class_weight=class_weights,

    callbacks=make_callbacks(
        "ANN"
    ),

    verbose=1
)
```

```python
plt.figure(figsize=(12,4))

plt.subplot(1,2,1)

plt.plot(
    ann_history.history["accuracy"],
    label="Train"
)

plt.plot(
    ann_history.history["val_accuracy"],
    label="Validation"
)

plt.title("ANN Accuracy")
plt.legend()

plt.subplot(1,2,2)

plt.plot(
    ann_history.history["loss"],
    label="Train"
)

plt.plot(
    ann_history.history["val_loss"],
    label="Validation"
)

plt.title("ANN Loss")
plt.legend()

plt.show()
```

```python
yt_ann, yp_ann = eval_cls_model(
    ann_model,
    te128,
    "ANN"
)

all_histories["ANN"] = ann_history.history
```

```python
cm = confusion_matrix(
    yt_ann,
    yp_ann
)

plt.figure(figsize=(6,5))

sns.heatmap(
    cm,
    annot=True,
    fmt="d",
    cmap="Blues",
    xticklabels=CLASSES,
    yticklabels=CLASSES
)

plt.title("ANN Confusion Matrix")

plt.xlabel("Predicted")
plt.ylabel("Actual")

plt.show()
```

# 9. Model 2 – Convolutional Neural Network (CNN) with Sigmoid Activation

This model introduces Convolutional Neural Networks (CNNs) for automated feature extraction from chest X-ray images. The **Sigmoid** activation function is used in all hidden layers to evaluate its performance for multi-class Tuberculosis classification.

**Key Features**
- Three convolutional layers
- Max pooling for feature reduction
- Global Average Pooling
- Dropout regularization
- Softmax output layer

```python
cnn_sigmoid = tf.keras.Sequential([

    layers.Conv2D(
        32,
        (3,3),
        activation="sigmoid",
        padding="same",
        input_shape=(IMG_SIZE[0], IMG_SIZE[1], 3)
    ),
    layers.MaxPooling2D((2,2)),

    layers.Conv2D(
        64,
        (3,3),
        activation="sigmoid",
        padding="same"
    ),
    layers.MaxPooling2D((2,2)),

    layers.Conv2D(
        128,
        (3,3),
        activation="sigmoid",
        padding="same"
    ),
    layers.MaxPooling2D((2,2)),

    layers.GlobalAveragePooling2D(),

    layers.Dense(
        128,
        activation="sigmoid"
    ),

    layers.Dropout(0.3),

    layers.Dense(
        NUM_CLASSES,
        activation="softmax"
    )
])

cnn_sigmoid.compile(
    optimizer="adam",
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

cnn_sigmoid.summary()
```

```python
cnn_sigmoid_history = cnn_sigmoid.fit(

    tr128,

    validation_data=v128,

    epochs=EPOCHS,

    class_weight=class_weights,

    callbacks=make_callbacks(
        "CNN_SIGMOID"
    ),

    verbose=1
)
```

```python
plt.figure(figsize=(12,4))

plt.subplot(1,2,1)

plt.plot(
    cnn_sigmoid_history.history["accuracy"],
    label="Train"
)

plt.plot(
    cnn_sigmoid_history.history["val_accuracy"],
    label="Validation"
)

plt.title("CNN-Sigmoid Accuracy")
plt.legend()

plt.subplot(1,2,2)

plt.plot(
    cnn_sigmoid_history.history["loss"],
    label="Train"
)

plt.plot(
    cnn_sigmoid_history.history["val_loss"],
    label="Validation"
)

plt.title("CNN-Sigmoid Loss")
plt.legend()

plt.show()
```

```python
yt_sigmoid, yp_sigmoid = eval_cls_model(
    cnn_sigmoid,
    te128,
    "CNN-Sigmoid"
)

all_histories["CNN-Sigmoid"] = (
    cnn_sigmoid_history.history
)

all_results["CNN-Sigmoid"] = (
    yt_sigmoid,
    yp_sigmoid
)
```

# 10. Model 3 – Convolutional Neural Network (CNN) with Tanh Activation

This model uses the same CNN architecture as the previous model while replacing the **Sigmoid** activation function with **Tanh**. Tanh is zero-centered and often provides better gradient flow, making it a common alternative for deep neural networks.

**Key Features**
- Three convolutional layers
- Max pooling for feature reduction
- Global Average Pooling
- Dropout regularization
- Softmax output layer

# 11. Model 4 – Convolutional Neural Network (CNN) with ReLU Activation

- Global Average Pooling
- Dropout regularization
- Softmax output layer

# 12. EfficientNetB0

EfficientNetB0 is a transfer learning model pretrained on ImageNet.

Key Features:
- Compound scaling
- Lower parameter count
- Strong feature extraction
- Widely used in medical imaging

Objective:
Classify Chest X-Ray images into:
1. Healthy
2. Sick
3. Active TB

```python
from tensorflow.keras.applications.efficientnet import preprocess_input

eff_train_gen = ImageDataGenerator(
    preprocessing_function=preprocess_input,
    rotation_range=15,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.1,
    horizontal_flip=True
)

eff_val_gen = ImageDataGenerator(
    preprocessing_function=preprocess_input
)

tr_eff = eff_train_gen.flow_from_dataframe(
    train_df,
    x_col='image_path',
    y_col='label',
    target_size=(224,224),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=True
)

v_eff = eff_val_gen.flow_from_dataframe(
    val_df,
    x_col='image_path',
    y_col='label',
    target_size=(224,224),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=False
)

te_eff = eff_val_gen.flow_from_dataframe(
    test_df,
    x_col='image_path',
    y_col='label',
    target_size=(224,224),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=False
)

print("EfficientNet Generators Ready")
```

```python
from tensorflow.keras.applications import EfficientNetB0

base_model = EfficientNetB0(
    weights='imagenet',
    include_top=False,
    input_shape=(224,224,3)
)

base_model.trainable = False

inputs = tf.keras.Input(shape=(224,224,3))

x = base_model(inputs, training=False)

x = layers.GlobalAveragePooling2D()(x)

x = layers.BatchNormalization()(x)

x = layers.Dropout(0.5)(x)

x = layers.Dense(
    512,
    activation='relu'
)(x)

x = layers.Dropout(0.3)(x)

outputs = layers.Dense(
    NUM_CLASSES,
    activation='softmax'
)(x)

effnet = tf.keras.Model(
    inputs,
    outputs
)

effnet.compile(
    optimizer=tf.keras.optimizers.Adam(
        learning_rate=1e-3
    ),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

effnet.summary()
```

```python
print("Training EfficientNetB0...")

hist_eff = effnet.fit(
    tr_eff,
    validation_data=v_eff,
    epochs=15,
    class_weight=class_weights,
    callbacks=make_callbacks(
        "EfficientNetB0",
        patience=5
    ),
    verbose=1
)
```
